```python
import glob
val_fnames = list(docs.keys())

#
```

```python
val_texts = {os.path.basename(os.path.splitext(x)[0]): x for x in glob.glob(os.path.join(DATADIR, 'corpus', '*'))
            if os.path.basename(os.path.splitext(x)[0]) in val_fnames}
val_texts

val_xmls = {os.path.basename(x).split('.')[0]: x for x in glob.glob(os.path.join(DATADIR, 'annotations', '*.bioc.xml'))
            if os.path.basename(x).split('.')[0] in val_fnames}
len(val_xmls)
```

```python
assert len(val_texts) == len(val_xmls) == len(val_fnames)
```

```python
import shutil
for f in val_fnames:
    text = val_texts[f]
    assert os.path.exists(text)
    xml = val_xmls[f]
    assert os.path.exists(xml)
    
    shutil.copy(text, os.path.join(OUTDIR, 'corpus', os.path.basename(text)))
    shutil.copy(xml, os.path.join(OUTDIR, 'annotations', os.path.basename(xml)))
```

```python
doc = docs['10_633']
relat = doc.get_relations()[0]
relat
anno1 = relat.annotation_1
anno2 = relat.annotation_2
```

```python
import bioc
writer = bioc.BioCWriter()
writer.collection = bioc.BioCCollection()
collection = writer.collection
document = bioc.BioCDocument()
document.id = doc.file_name

passage = bioc.BioCPassage()

anno = bioc.bioc_annotation.BioCAnnotation()
anno.id = anno1.id
anno.put_infon('type', anno1.type)

loc = bioc.bioc_location.BioCLocation()
loc.offset = str(anno1.start_index)
loc.length = str(len(anno1.text))
anno.add_location(loc)
#anno.
anno.text = anno1.text

relation = bioc.bioc_relation.BioCRelation()
relation.id = relat.id
relation.put_infon('type', relat.type)
node1 = bioc.bioc_node.BioCNode()
node1.refid = anno1.id
node1.role = 'annotation 1'
relation.add_node(node1)

node2 = bioc.bioc_node.BioCNode()
node2.refid = anno2.id
node2.role = 'annotation 2'
relation.add_node(node2)

passage.add_relation(relation)



passage.offset = '0'
passage.add_annotation(anno)
document.add_passage(passage)
collection.add_document(document)

print(writer)
```
